```python
#Numerical Analysis
numerical_cols = ['Estimated Income','Superannuation Savings','Credit Card Balance', 'Bank Loans','Bank Deposits', 'Checking Accounts', 'Saving Accounts','Foreign Currency Account', 'Business Lending']
plt.figure(figsize = (20,10))
for i,col in enumerate(numerical_cols):
  plt.subplot(4,3,i+1)
  sns.histplot(df[col],kde = True)
  plt.title(f'Distribution of {col}')
plt.tight_layout()
plt.show()
```

The histograms in the image display the distribution of financial variables. Most variables, such as Bank Deposits, Bank Loans, Saving Accounts, Checking Accounts, and Business Lending, show right-skewed distributions, indicating that a majority of individuals have lower values, with fewer individuals holding very high amounts. Estimated Income and Superannuation Savings are also skewed but appear more spread out, suggesting moderate variability. Credit Card Balance and Foreign Currency Account similarly exhibit right-skewed patterns. Overall, the distributions reveal that wealth and financial products are unevenly distributed, with most people having modest amounts and a smaller group holding significantly higher values.

```python
#Heatmaps
numerical_cols = ['Estimated Income','Superannuation Savings','Credit Card Balance', 'Bank Loans','Bank Deposits', 'Checking Accounts', 'Saving Accounts','Foreign Currency Account', 'Business Lending']
correlation_matrix = df[numerical_cols].corr()
plt.figure(figsize = (10,8))
sns.heatmap(correlation_matrix,annot = True,cmap = 'crest',fmt = '.2f')
plt.title('Correlation Matrix')
plt.show()
```

The correlation matrix shows strong positive relationships between bank-related variables, particularly between bank deposits and checking accounts (0.84) and bank deposits and saving accounts (0.75), indicating that individuals who maintain higher deposits often do so across multiple account types. Moderate correlations exist between credit card balance, bank loans, and business lending, suggesting a link between credit usage and lending activity. In contrast, estimated income and superannuation savings have weaker correlations with other variables, implying that they function more independently of daily banking behavior. Overall, financial product usage tends to cluster, especially around deposits and account balances.
